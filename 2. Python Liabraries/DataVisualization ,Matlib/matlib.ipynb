**MATLIB**

 What is Matplotlib?

Matplotlib is a 2D plotting library that is widely used for creating static, animated, and interactive visualizations in Python.
It allows you to create a wide range of plots and charts, including line plots, bar charts, histograms, scatter plots, and more.

* A Python library for visualization
* Works with NumPy & Pandas
* Creates static, animated, and interactive plots

Graph Types in Matplotlib – Functions & Use Cases

* Line Plot ⇒ plt.plot() ⇒ Trends over time

* Bar Chart ⇒ plt.bar() ⇒ Comparing categories

* Histogram ⇒ plt.hist() ⇒ Distribution of data

* Scatter Plot ⇒ plt.scatter() ⇒ Relationship between variables

* Box Plot ⇒ plt.boxplot() ⇒ Detecting outliers

* Pie Chart ⇒ plt.pie() ⇒ Showing proportions



# LINE PLOT

* A continuous line connecting data points
* Used to show trends, patterns, and changes over time

**When to Use?**

✅ When analyzing time-series
data

✅ When tracking continuous
data changes

**When NOT to Use?**

❌ When comparing categories (Use
bar chart instead)

❌ When dealing with discrete
variables

Basic Plotting with Matplotlib

```python
import matplotlib.pyplot as plt
```

 Simple Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.title('Simple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

```python
# x = [1, 2, 3, 4, 5]
# y = [10, 20, 15, 25, 30]
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, marker='o', linestyle='--', color='r')
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Line Plot Example")
plt.show()
```

* plt.plot(x, y): This plots the line connecting the points defined by x and y.

* plt.title(): Adds a title to the plot.

* plt.xlabel(), plt.ylabel(): Labels the X and Y axes.

* plt.grid(True): Adds a grid for better readability.

* plt.show(): Displays the plot.

# BAR CHART – COMPARING CATEGORIES

* Represents categorical data with rectangular bars
* Bar height represents value/count

**When to Use?**

✅ When comparing categories
or groups

✅ When showing discrete data
distributions

**When NOT to Use?**

❌When visualizing continuous trends
(Use line plot instead)

❌ When displaying too many categories
(Use grouped bar charts or different
visualization)

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
categories = ["A","B","C","D"]
values = [10, 15, 7, 12]
plt.bar(categories, values, color='blue')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

# SCATTER PLOT – RELATIONSHIPS BETWEEN VARIABLES

* Represents individual data points
* Helps to identify correlations & patterns

**When to Use?**

✅When analyzing relationships
between two continuous variables

✅ When checking for clusters or
trends

**When NOT to Use?**

❌ When one variable is categorical (Use
bar chart instead)

❌ When there’s too much overlapping
data (Use hexbin plot)

```python
x = [1, 2, 3, 4, 5]
y = [5, 6, 7, 8, 9]

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
x = [10, 20, 30, 40, 50]
y = [5, 15, 25, 35, 45]
plt.scatter(x, y, color='red')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

# HISTOGRAM – DATA DISTRIBUTION

* Shows frequency distribution of a dataset
* Bins group data into ranges

**When to Use?**

✅ When analyzing distribution
& shape of data

✅ When checking skewness or
normality

**When NOT to Use?**

❌ When comparing individual values
(Use bar chart)

❌ When dealing with categorical data

```python
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=20, color='green'
, alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

# BOX PLOT – DETECTING OUTLIERS

* Displays data distribution, median, and outliers
* Uses quartiles to summarize spread

**When to Use?**

✅ When detecting outliers &
anomalies

✅ When comparing multiple
distributions

**When NOT to Use?**

❌ When visualizing exact values (Use
scatter plot)

❌ When comparing few data points

```python
data = [7, 8, 5, 12, 15, 9, 10, 18, 25, 30]
plt.boxplot(data)
plt.title("Box Plot Example")
plt.show()
```

# PIE CHART – SHOWING PROPORTIONS

* Represents parts of a whole
* Each slice = percentage of total

**When to Use?**

✅ When showing proportions &
percentages

✅ When comparing a few
categories

**When NOT to Use?**

❌ When categories exceed 5-6 (Use bar
chart instead)

❌ When values are similar in size
(Difficult to differentiate)

```python
labels = ["A","B","C","D"]
sizes = [30, 20, 25, 25]
plt.pie(sizes, labels=labels, autopct="%1.1f%%",colors=["red","blue","green","yellow"])
plt.title("Pie Chart Example")
plt.show()
```

# 📊 Step 4: Customizing Plots

* Adding Colors, Line Styles, and Markers
* You can customize your plots in various ways. For example, adding color and markers.

```python
x = [1, 2, 3, 4, 5]
y = [5, 6, 7, 8, 9]

plt.plot(x, y, color='green', linestyle='--', marker='o', markersize=8)
plt.title('Customized Plot')
plt.show()
```

# Multiple Plots in One Figure

You can plot multiple graphs in one figure using subplot().

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.subplot(1, 2, 1)  # (rows, columns, index)
plt.plot(x, y, 'r-')
plt.title('Line Plot')

y2 = [5, 4, 3, 2, 1]
plt.subplot(1, 2, 2)
plt.plot(x, y2, 'b-')
plt.title('Line Plot 2')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

**Interview Qs Based on Matplotlib**

**Q1: What are the basic types of plots in Matplotlib?**

Line Plot: Best for continuous data.

Bar Plot: For categorical data comparisons.

Scatter Plot: Used for showing relationships between two continuous variables.

Histogram: To show the distribution of a dataset.

**Q2: How do you add labels and titles in a Matplotlib plot?**

You can use plt.title(), plt.xlabel(), and plt.ylabel() to add titles and axis labels.


**Q3: What is the use of plt.tight_layout()?**

It adjusts the spacing between subplots to prevent label overlap.


**Q4: How do you plot multiple lines on the same plot?**

You can call plt.plot() multiple times before calling plt.show() to overlay multiple lines on a single graph.
